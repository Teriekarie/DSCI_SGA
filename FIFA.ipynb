```python
#Import Dependencies
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#load csv File into DataFrame
df = pd.read_csv('/content/drive/MyDrive/STUTERN 1.3 ASSIGNMENT/module8/WorldCupMatches.csv')
df
```

```python
#Check missing values and dropping them
df.isnull().sum()
df.dropna( inplace = True)
df.head()
```

```python
#To visualize the total number of goals scored in each match
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
df.head()
```

```python
# To visualize the total number of goals scored in each match
res = sns.kdeplot(data=df['Total Goals'], color='r', shade=True)
plt.title('KDE Plot For The Total Goals')
plt.show()
```

```python
#Create a bar chart visualizing how many goals were scored each year the World Cup was held between 1930-2014.
sns.set_style('whitegrid')
sns.set_context('poster', font_scale=0.4)
f, ax = plt.subplots(figsize=(12, 7))
ax = sns.barplot(data=df, x='Year', y='Total Goals')
ax.set_title('Goals every year')
plt.show()
```

```python
#Visualize the columns Year and Total Goals as a bar chart.
fig= plt.subplots(figsize=(12,7))
ax = sns.barplot(data=df, x="Year", y="Total Goals", ci=None, estimator=sum)
ax.set_title("Total Goals Scored Yearly".upper())  
plt.show()
```

```python
#To visualize the distribution of the average goals for each year.
fig= plt.figure(figsize=(12,7))
sns.boxplot(data=df,x='Year', y='Total Goals')
plt.title('Average goals for each year')
plt.show()
```
