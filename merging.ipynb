```python
# create an inner merge between the orders table and products table
df_orders_products = df_order.merge(df_products, how='inner', on='product_id')
df_orders_products.head()
```

This merge method combines the similar columns from the orders dataframe and product dataframe and adds the remaining columns in the product dataframe

```python
# create an inner merge between the orders table and customers table
df_orders_customers = df_order.merge(df_customers, how='inner', on='customer_id')
df_orders_customers.head()
```

This merge method combines the similar columns from the orders dataframe and customers dataframe and adds the remaining columns in the customers dataframe

```python
#Use the resulting table to perform another inner merge operation with the customers table.
df_all = df_orders_products.merge(df_customers, how='inner')
df_all.head()
```

This merge combines the similar columns left and right.

```python
# Perform an inner merge operation between customers and orders tables, will be merging customer_id
df_customers_orders = df_customers.merge(df_order, how='inner', on='customer_id')
df_customers_orders.head()
```

This showed all the all the columns on both dataframes.

```python
#Perform an outer merge between customers and products tables.
df_customers_products = pd.merge(df_customers, df_products, how ='outer')
df_customers_products()
```

```python
#Perform an outer merge between customers and products tables. using left_index,right_index
df_customers_products = pd.merge(df_customers,df_products,left_index=True,right_index=True,how ='outer')
df_customers_products.head()
```

All the matching and non matching records are available in resultant dataset from both data frames

```python
#Perform a left merge between orders and customers tables. picking customer_id for merge
df_orders_customers = pd.merge(df_order,df_customers, how='left', on=['customer_id'])
df_orders_customers
```

Because the rows of the customer dataframe does not tally with that of the order dataframe, NaN was generated to fill the gaps

```python
# perform the right merge between customers and products dataframes
df_products_customers = pd.merge(df_products, df_customers,left_index=True,right_index=True,how = 'right')
df_products_customers.head()
```

```python
#Use pandas concatenation method to join the three datasets together, concatenating along columns
df_concat = pd.concat([df_order, df_customers,df_products], axis=1)
df_concat
```
