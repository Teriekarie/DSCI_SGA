```python
#Import Dependencies
import numpy as np 
import pandas as pd
```

```python
#Load csv File
movies= pd.read_csv('/content/drive/MyDrive/STUTERN 1.3 ASSIGNMENT/module8/movies.csv')
movies
```

Formating the column names

```python
movies.columns = movies.columns.str.upper()
movies.columns
```

Changing the ONE-LINE colunmn name to SYNOPSIS

```python
movies.rename= movies.rename(columns = {'ONE-LINE':'SYNOPSIS'}, inplace=True)
movies.rename
```

```python
#Check for Missing Values in the dataframe
movies.isnull().any().any()
```

```python
#Check for Missing Values in each column of the Dataset
movies.isnull().any()
```

This shows the column with the empty cells

```python
#Check for the total missing values in each column
movies.isnull().sum()
```

Cleaning the empty columns.
Lets replace the rating, votes and runtime with the average of each column and drop the year and genre as such column can not be assumed.

```python
#converting the datatype of votes to be able to get the mean
#Removing Comma from 'VOTES' column
movies['VOTES'] = movies['VOTES'].astype(str)
movies['VOTES'] = [x.replace(',','') for x in movies["VOTES"]]
# convert "VOTES" from string to float
movies['VOTES'] = movies['VOTES'].astype(float)
#correct data types of columns
movies.dtypes
```

```python
# replacing the column
movies_runtime= movies['RUNTIME'].fillna(movies['RUNTIME'].mean(), inplace=True)

#fill the rating columns
movies_rating= movies['RATING'].fillna(movies['RATING'].mean(), inplace=True)

#fill the votes column
movies_votes = movies['VOTES'].fillna(movies['VOTES'].mean(), inplace=True)

# Fill up empty spaces  for 'Gross' column with 0
movies['GROSS'].fillna(0, inplace=True)

# Dropping rows with missing values in Year and Genre
movies.dropna(axis='index', how='any', subset=['YEAR', 'GENRE'], inplace=True)

movies
```

```python
movies.isnull().sum()
```

Now we have all the rows in the dataframe filled up

```python
#Checking for duplicates
movies.duplicated().any()
```

```python
# removing the duplicates
movies.drop_duplicates(inplace= True)
movies.duplicated().sum()
```

Final touch to ensure the dataframe is clean and ready for use

```python
#More Data cleaning to make the dataset neat
#Removing the next line(\n) value from the columns
movies["GENRE"] = movies["GENRE"].astype(str)
movies["GENRE"] = [x.replace('\n','') for x in movies["GENRE"]]


movies["SYNOPSIS"] = movies["SYNOPSIS"].astype(str)
movies["SYNOPSIS"] = [x.replace('\n','') for x in movies["SYNOPSIS"]]

movies["STARS"] = movies["STARS"].astype(str)
movies["STARS"] = [x.replace('\n','') for x in movies["STARS"]]

movies
```
